```python
import pandas as pd
import numpy as np
```

```python
pd.set_option('display.max_columns', None)
db=pd.read_csv("database.csv", sep=";")
```

```python
db=db[(db['A6']=='Twitch') | (db['A6']=='Youtube')]
```

### Platform related variables

#### Channel topic

```python
db.loc[db['A6']=='Twitch', 'topic'] = db['A5[A24]']
db.loc[db['A6']=='Youtube', 'topic'] = db['A5[A25]']
db['topic'].unique()
```

```python
db['topic'].value_counts()
```

```python
db['topic']=db['topic'].replace('Lifestyle : mode, beauté', 'Autre')
db['topic']=db['topic'].replace('Musique', 'Autre')
db['topic']=db['topic'].replace('Cuisine', 'Autre')
```

```python
db['topic'].value_counts()
```

#### Production period - seniority on platform

```python
db.loc[db['A6']=='Twitch', 'duration'] = db['A3[A24]']
db.loc[db['A6']=='Youtube', 'duration'] = db['A3[A25]']
db['duration'].value_counts()
```

```python
db.loc[db['duration']=="Depuis moins d'un an.", 'duration']= 'Less than 2 years'
db.loc[db['duration']=='Entre un et deux ans.', 'duration']= 'Less than 2 years'

db.loc[db['duration']=='Entre deux et quatres ans', 'duration']= 'Between 2 and 4 years'
db.loc[db['duration']=='Depuis cinq ans et plus.', 'duration']= '5 years and more'

db['duration'].value_counts()
```

#### Number of followers

```python
db.loc[db['A6']=='Twitch', 'follow'] = db['A4[A24]']
db.loc[db['A6']=='Youtube', 'follow'] = db['A4[A25]']

db.loc[db['follow']=='Moins de 1 000 personnes.', 'following'] = 'Less than 1 000 followers'
db.loc[db['follow']=='Entre 1 000 et 10 000 personnes.', 'following'] = 'Between 1 000 and 10 000 followers'

db.loc[db['follow']=='Entre 10 000 et 100 000 personnes.', 'following'] = 'Between 10 000 and 100 000 followers'
db.loc[db['follow']=='Plus de 100 000 personnes.', 'following'] = 'More than 100 000 followers'
db['following'].value_counts()
```
