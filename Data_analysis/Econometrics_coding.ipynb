```python
import pandas as pd
import numpy as np
```

```python
pd.set_option('display.max_columns', None)
```

```python
#Loading DB MCA
db=pd.read_csv('db_mca.csv', sep=';')
db
```

```python
#Duplicating db for coding variable
db_eco=db.copy()
db_eco
```

# Coding variables

## Production variables

```python
#frequency
db_eco['frequency']=db_eco['freq'].copy()
db_eco['frequency']=db_eco['frequency'].replace("Rare",0)
db_eco['frequency']=db_eco['frequency'].replace("Monthly",1)
db_eco['frequency']=db_eco['frequency'].replace("Weekly",2)

db_eco['freq_weekly']=np.where(db_eco['freq']=="Weekly",1,0)
db_eco['freq_monthly']=np.where(db_eco['freq']=="Monthly",1,0)
db_eco['freq_rare']=np.where(db_eco['freq']=="Rare",1,0)

#format
db_eco['form']=db_eco['format'].copy()
db_eco['form']=db_eco['form'].replace("VOD_Live",0)
db_eco['form']=db_eco['form'].replace("VOD_only",1)
db_eco['form']=db_eco['form'].replace("Live_only",2)

db_eco['format_live']=np.where(db_eco['format']=="Live_only",1,0)
db_eco['format_VOD']=np.where(db_eco['format']=="VOD_only",1,0)
db_eco['format_VOD_live']=np.where(db_eco['format']=="VOD_Live",1,0)

#Help in production
db_eco['help_prod']=db_eco['help_prod'].replace('No', 0)
db_eco['help_prod']=db_eco['help_prod'].replace('Yes',1)

#Help in management
db_eco['help_manag']=db_eco['help_manag'].replace('No', 0)
db_eco['help_manag']=db_eco['help_manag'].replace('Yes',1)

#Multihoming
db_eco['MH']=db_eco['MH'].replace(False, 0)
db_eco['MH']=db_eco['MH'].replace(True,1)
```

```python
db_eco
```

## Monetization variables

```python
#Revenue income
# rev_ad
db_eco['rev_ad']=db_eco['rev_ad'].replace('No', 0)
db_eco['rev_ad']=db_eco['rev_ad'].replace('Yes',1)

# rev_subs
db_eco['rev_subs']=db_eco['rev_subs'].replace('No', 0)
db_eco['rev_subs']=db_eco['rev_subs'].replace('Yes',1)

# rev_sponso
db_eco['rev_sponso']=db_eco['rev_sponso'].replace('No', 0)
db_eco['rev_sponso']=db_eco['rev_sponso'].replace('Yes',1)

# rev_donation
db_eco['rev_don']=db_eco['rev_don'].replace('No', 0)
db_eco['rev_don']=db_eco['rev_don'].replace('Yes',1)

# rev_other
db_eco['rev_other']=db_eco['rev_other'].replace('No', 0)
db_eco['rev_other']=db_eco['rev_other'].replace('Yes',1)

#revenue_share
db_eco['share_rev_low']=np.where(db_eco['rev_share']=="Less than 10%",1,0)
db_eco['share_rev_medium']=np.where(db_eco['rev_share']=="Between 11% and 50%",1,0)
db_eco['share_rev_high']=np.where(db_eco['rev_share']=="More than 50%",1,0)

db_eco['rev_share_cod']=db_eco['rev_share'].copy()
db_eco['rev_share_cod']=db_eco['rev_share_cod'].replace("Less than 10%", 0)
db_eco['rev_share_cod']=db_eco['rev_share_cod'].replace("Between 11% and 50%", 1)
db_eco['rev_share_cod']=db_eco['rev_share_cod'].replace("More than 50%", 2)

#2021 income
db_eco['low_income']=np.where(db_eco['rev_2021']=="Less than 15000€",1,0)
db_eco['medium_income']=np.where(db_eco['rev_2021']=="Between 15000€ and 30000€",1,0)
db_eco['high_income']=np.where(db_eco['rev_2021']=="More than 30000€",1,0)
```

```python
db_eco
```

## Additional variables

```python
#Number of followers
db_eco['fol']=db_eco['follow'].copy()
db_eco['fol']=db_eco['fol'].replace("Less than 10 000 followers",0)
db_eco['fol']=db_eco['fol'].replace("More than 10 000 followers",1)

db_eco['low_nb_follower']=np.where(db_eco['follow']=="Less than 10 000 followers",1,0)
db_eco['high_nb_follower']=np.where(db_eco['follow']=="More than 10 000 followers",1,0)

#Production period
db_eco['prodp']=db_eco['duration'].copy()
db_eco['prodp']=db_eco['prodp'].replace("Less than 2 years",0)
db_eco['prodp']=db_eco['prodp'].replace("Between 2 and 4 years",1)
db_eco['prodp']=db_eco['prodp'].replace("5 years and more",2)

db_eco['small_prod_period']=np.where(db_eco['duration']=="Less than 2 years",1,0)
db_eco['medium_prod_period']=np.where(db_eco['duration']=="Between 2 and 4 years",1,0)
db_eco['high_prod_period']=np.where(db_eco['duration']=="5 years and more",1,0)

#Platform creator program membership
db_eco['creator_prog']=db_eco['creator_prog'].replace('No', 0)
db_eco['creator_prog']=db_eco['creator_prog'].replace('Yes',1)
```

```python
#Main platform
db_eco['A6']=db_eco['A6'].replace('Twitch', 0)
db_eco['A6']=db_eco['A6'].replace('Youtube', 1)
```

```python
db_eco
```

# Creating dataframe for regressions

```python
db_eco=db_eco.drop(['Unnamed: 0','rev_share', 'rev_2021'], axis=1)
```

```python
db_eco.to_csv('db_eco.csv', sep=';')
```
