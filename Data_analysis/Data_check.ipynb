# Test 1 - Cohérence des revenus par plateforme

Pas de revenus publicitaires : Facebook, Instagram et TikTok

Pas de revenus d’abonnement : Facebook, Instagram et TikTok

```python
#Check des revenus publicitaires

#Facebook
condition_trpf=[(db['A6']=='Facebook') & (db['B13[RB131]']=='Oui')]
value_trpf=['True']
db['T1[RPF]']=np.select(condition_trpf, value_trpf)
db['T1[RPF]']=db['T1[RPF]'].replace("0", 'False')

#Instagram
condition_trpi=[(db['A6']=='Instagram') & (db['B13[RB131]']=='Oui')]
value_trpi=['True']
db['T1[RPI]']=np.select(condition_trpi, value_trpi)
db['T1[RPI]']=db['T1[RPI]'].replace("0", 'False')

#TikTok
condition_trpt=[(db['A6']=='TikTok') & (db['B13[RB131]']=='Oui')]
value_trpt=['True']
db['T1[RPT]']=np.select(condition_trpt, value_trpt)
db['T1[RPT]']=db['T1[RPT]'].replace("0", 'False')
```
