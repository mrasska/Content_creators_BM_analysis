## Platform combinaison

### Total combinaison

```python
#Idea = db['combo'][1]=='Main_second_second'

#For each column of platform production, change 'Oui' by the name of the platform and 'Non' by N/A
db['prod_fb']=db['prod_fb'].replace("Oui", "Facebook")

db['prod_insta']=db['prod_insta'].replace("Oui", "Instagram")

db['prod_tk']=db['prod_tk'].replace("Oui", "TikTok")

db['prod_tw']=db['prod_tw'].replace("Oui", "Twitch")

db['prod_yt']=db['prod_yt'].replace("Oui", "YouTube")
```

```python
db['combo'] = db[['prod_fb', 'prod_insta', 'prod_tk', 'prod_tw', 'prod_yt']].agg('_'.join, axis=1)
```

```python
db['combo']=db['combo'].str.replace('Non_','')
db['combo']=db['combo'].str.replace('Non','')
```

```python
#dataframe with the count of platforms combo
df_combo=db.groupby(['A6','combo']).size().reset_index(name='counts')
df_combo.to_csv('platform_combo.csv', sep=';', encoding='utf-8')
df_combo
```
